## II. pyplot
1. process of pyplot
![process of pyplot](pics/pyplot_struct.png)

```python
import matplotlib.pyplot as plt
from matplotlib.axes import Axes, Subplot
import numpy as np
```

2. Create draw_board:<br>
1). plt.figure，主要接收一个元组作为figsize参数设置图形大小，返回一个figure对象用于提供画板;<br>

```python
fig, axes = plt.subplots(2, 2)
print('fig: ', fig)
```

2). plt.axes，接收一个figure或在当前画板上添加一个子图，返回该axes对象，并将其设置为"当前"图，缺省时会在绘图前自动添加;<br>

```python
fig, axes = plt.subplots(2, 2)
print('axes:', axes)
plt.plot([1, 2])
```

3). plt.subplot，主要接收3个数字或1个3位数（自动解析成3个数字，要求解析后数值合理）作为子图的行数、列数和当前子图索引，索引从1开始（与MATLAB保存一致），返回一个axes对象用于绘图操作。这里，可以理解成是先隐式执行了plt.figure，然后在创建的figure对象上添加子图，并返回当前子图实例;<br>plt.subplots，主要接收一个行数nrows和列数ncols作为参数（不含第三个数字），创建一个figure对象和相应数量的axes对象，同时返回该figure对象和axes对象嵌套列表，并默认选择最后一个子图作为"当前"图<br>

```python
ax = plt.subplot(221) # 2行2列的大图，第一张子图
ax.plot([2, 1])
ax1 = plt.subplot(224) # 2行2列的大图，第四张子图
plt.plot([1, 2])
```

3. Types:<br>
1). plot，折线图或点图，实际是调用了line模块下的Line2D图表接口<br>
2). scatter，散点图，常用于表述两组数据间的分布关系，也可由特殊形式下的plot实现<br>
3). bar/barh，条形图或柱状图，常用于表达一组离散数据的大小关系，比如一年内每个月的销售额数据；默认竖直条形图，可选barh绘制水平条形图<br>
4). hist，直方图，形式上与条形图很像，但表达意义却完全不同：直方图用于统计一组连续数据的分区间分布情况，比如有1000个正态分布的随机抽样，那么其直方图应该是大致满足钟型分布；条形图主要是适用于一组离散标签下的数量对比<br>
5). pie，饼图，主要用于表达构成或比例关系，一般适用于少量对比<br>
imshow，显示图像，根据像素点数据完成绘图并显示<br>

4. Legend:<br>
![legend](pics/pyplot_legend.png)
1). title，设置图表标题<br>
2). axis/xlim/ylim，设置相应坐标轴范围，其中axis是对后xlim和ylim的集成，接受4个参数分别作为x和y轴的范围参数<br>
3). grid，添加图表网格线<br>
4). legend，在图表中添加label图例参数后，通过legend进行显示<br>
5). xlabel/ylabel，分别用于设置x、y轴标题<br>
6). xticks/yticks，分别用于自定义坐标轴刻度显示<br>
7). text/arrow/annotation，分别在图例指定位置添加文字、箭头和标记，一般很少用<br>

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, label = 'y1 = sin(x)')
plt.plot(x, y2, label = 'y2 = cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.axis([0, 10, -1, 1])
plt.legend(loc=1)
plt.xticks([0, 3.14, 6.28, 9.42]) #, ['0', '$\pi$', '$\2pi$', '$3\pi$']
plt.yticks([-1, 0, 1])
```

5. 接口对比: <br>
plt.axes()——fig.add_axes()
plt.subplot()——fig.add_subplot()
plt.GridSped()——fig.add_gridspec()
plt.xlabel()——axes.set_xlabel()
plt.ylabel()——axes.set_ylabel()
plt.xlim()——axes.set_xlim()
plt.ylim()——axes.set_ylim()
plt.title()——axes.set_title()

```python
fig, axes = plt.subplots(2, 1)

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

ax = plt.subplot(211)
ax.set_xlabel('x')
ax.set_ylabel('y1')
plt.plot(x, y1, label = 'y1 = sin(x)')

ax1 = plt.subplot(212)
ax1.set_xlabel('x')
ax1.set_xlabel('y2')
plt.plot(x, y2, label = 'y2 = cos(x)')
```
